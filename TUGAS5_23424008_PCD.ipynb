1.Lowpass Filter

```python

from google.colab import files
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import io

# Pilih file dari komputer kamu
uploaded = files.upload()

# Mengambil nama file yang diupload dan membacanya
for filename in uploaded.keys():
    img_data = uploaded[filename]
    img = Image.open(io.BytesIO(img_data)).convert('L') # Ubah ke grayscale
    img = np.array(img)

print("Foto berhasil diupload!")
# Menggunakan Gaussian Blur sebagai LPF
# (15, 15) adalah ukuran kernel, semakin besar semakin blur
lpf_img = cv2.GaussianBlur(img, (15, 15), 0)

plt.imshow(lpf_img, cmap='gray')
plt.title("Hasil Low-Pass Filter")
plt.show()
```

2. Highpass Filter

```python
from google.colab import files
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import io

# Pilih file dari komputer kamu
uploaded = files.upload()

# Mengambil nama file yang diupload dan membacanya
for filename in uploaded.keys():
    img_data = uploaded[filename]
    img = Image.open(io.BytesIO(img_data)).convert('L') # Ubah ke grayscale
    img = np.array(img)

print("Foto berhasil diupload!")
# HPF menggunakan Laplacian
hpf_hasil = cv2.Laplacian(img, cv2.CV_64F)

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(122), plt.imshow(hpf_hasil, cmap='gray'), plt.title('High-Pass Filter')
plt.show()
```

3. Bandpass Filter

```python
from google.colab import files
import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import io

# Pilih file dari komputer kamu
uploaded = files.upload()

# Mengambil nama file yang diupload dan membacanya
for filename in uploaded.keys():
    img_data = uploaded[filename]
    img = Image.open(io.BytesIO(img_data)).convert('L') # Ubah ke grayscale
    img = np.array(img)

print("Foto berhasil diupload!")
# BPF menggunakan selisih dua Gaussian Blur (Difference of Gaussians)
blur1 = cv2.GaussianBlur(img, (5, 5), 0)
blur2 = cv2.GaussianBlur(img, (25, 25), 0)
bpf_hasil = blur1 - blur2

plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(122), plt.imshow(bpf_hasil, cmap='gray'), plt.title('Band-Pass Filter')
plt.show()
```
